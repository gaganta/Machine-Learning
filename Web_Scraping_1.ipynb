```python
import os
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://in.finance.yahoo.com/quote/000001.SS/history?p=000001.SS" #url where data is stored

response = requests.get(url) #pulling data to our environment
```

```python
response.content #lets see the data
```

```python
soup = BeautifulSoup(response.content, "html.parser") #sending data to cleaning process
```

```python
data = soup.find('table' , {'class' : 'W(100%) M(0)'}) #finding reg data
data = data.findAll('span')
```

```python
data
```

```python
dat = data[7: -3] #data sizing
dat = [x.text for x in dat] #taking only text from data
dat = np.array(dat).reshape(100,7)
```

```python
dat
```

```python
cols = data[:7]
cols = [col.text for col in cols]
cols
```

```python
df = pd.DataFrame(data = dat , columns = cols)
```
